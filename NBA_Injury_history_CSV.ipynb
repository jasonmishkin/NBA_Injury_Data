```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
```

```python
nba_injury_history_first_csv = pd.read_csv('Injury_History.csv')
nba_injury_history_first_csv
```

```python
# Connect to the database
engine = create_engine("mysql+pymysql://root:mishkin1@localhost/nba_injury")
```

```python
# Load the DataFrame into the database
nba_injury_history_first_csv.to_sql(name='nba_injury_history_1', con=engine, if_exists='replace', index=False)
```
