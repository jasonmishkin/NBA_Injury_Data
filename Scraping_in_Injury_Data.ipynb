# By Team

```python
def team_position_injury(url,x1,x2,team_position_injury):   
    # Create an URL object
    url = url
    # Open URL and pass to BeautifulSoup
    html = urlopen(url)
    stats_page = BeautifulSoup(html)

    
    # Collect table headers
    #column_headers = stats_page.findAll('tr')[0]
    #column_headers = [i.getText() for i in column_headers.findAll('th')]
    #print(column_headers)


    # Collect table rows
    rows = stats_page.findAll('tr')[x1:x2]
    # Get stats from each row
    injury_cash = []
    for i in range(len(rows)):
        injury_cash.append([col.getText() for col in rows[i].findAll('td')])

    Columns = [team_position_injury, 'Players','Games','Cash_Earned_While_Injured']
    df = pd.DataFrame(injury_cash, columns= Columns)
    return df
```

# By Player

```python
def player(url,x1,x2):   
    # Create an URL object
    url = url
    # Open URL and pass to BeautifulSoup
    html = urlopen(url)
    stats_page = BeautifulSoup(html)

    # Collect table headers
    #column_headers = stats_page.findAll('tr')[0]
    #column_headers = [i.getText() for i in column_headers.findAll('th')]
    #print(column_headers)

    # Collect table rows
    rows = stats_page.findAll('tr')[x1:x2]
    # Get stats from each row
    injury_cash = []
    for i in range(len(rows)):
        injury_cash.append([col.getText() for col in rows[i].findAll('td')])
        
    Columns = ['Name','Position','Team','Injury','Games','Cash_Earned_While_Injured']
    df = pd.DataFrame(injury_cash, columns= Columns)
    df = df.replace(r'\n',  ' ', regex=True)

    # Define a custom function to extract the name
    def extract_name(value):
        if value is not None:
        # Split the value by the space character
            values = value.split(' ')
        # Return the first value (the name)
            full_name = ' '.join(values[1:3])
            # Return the full name
            return full_name
        else:
        # Return a default value if the value is None
            return 'N/A'
    # Apply the custom function to the name column
    df['Name'] = df['Name'].apply(extract_name)
    
    return df
```

```python
def df_to_sql(dfq,table_name):
    # Connect to the database
    engine = create_engine("mysql+pymysql://root:mishkin1@localhost/nba_injury")
    # Load the DataFrame into the database
    dfq.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
```
